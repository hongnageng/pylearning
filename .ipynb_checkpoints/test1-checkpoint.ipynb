```python
a = "     hello world, this is a test"
b = a.replace(' ', '_')
print(b)
c = a.split(',')
print(c)
d = a.find('a') #寻找第一个出现a的index'
print(d)
e = a.strip()#去掉两段的空格,默认为空格，可指定
print(e)
f = e.rfind('a') #从右边开始找，找到第一个a的index
print(f)
```
